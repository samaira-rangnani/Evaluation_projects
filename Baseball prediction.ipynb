MACHINE LEARNING

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso,Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor
from sklearn.neighbors import KNeighborsRegressor
```

Building train test split

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.30)
```

```python
#LR
lr=LinearRegression()
lr.fit(x_train,y_train)
ypred=lr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#Lasso
ls=Lasso()
ls.fit(x_train,y_train)
ypred=ls.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#Ridge
rd=Ridge()
rd.fit(x_train,y_train)
ypred=rd.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#SVR
svr=SVR()
svr.fit(x_train,y_train)
ypred=svr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#RandomForest
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
ypred=rfr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#ExtraTrees
etr=ExtraTreesRegressor()
etr.fit(x_train,y_train)
ypred=etr.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#KNN
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
ypred=knn.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
From above r2 score we conclude that Lasso model providing the best
```

EVALUATING THE SCORE AGIN USING CROSS VAL SCORE

```python
model=[lr,ls,rd,svr,etr,rfr,knn]
for m in model:
    score=cross_val_score(m,x,y,cv=7).mean()
    print('Cross validation score of ',m, "is", score)
    print('Difference between r2 score and cross val score is',(r2_score(y_test,ypred)-score)*100)
    print("--"*100)
```

From above validation we observe that error between r2 score and cross val score is less for lasso model
Also Lasso model giving the best accuracy score for better prediction
Hence Selecting Lasso model as final model


LASSO MODEL FOR HYPER PARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV
params={'alpha': (np.logspace(-8, 8, 100))}
gcv=GridSearchCV(ls,params,cv=5)
gcv.fit(x_train,y_train)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
#lasso 
lasso=Lasso(alpha= 0.8302175681319752)
lasso.fit(x_train,y_train)
ypred=lasso.predict(x_test)
acc=r2_score(y_test,ypred)
print(acc)
print(mean_absolute_error(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```
