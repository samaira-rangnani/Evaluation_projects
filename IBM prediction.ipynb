MACHINE LEARNING

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,GradientBoostingClassifier,BaggingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
maxrs=0            ##Initializing randomstate
maxacc=0           ##Initialising accurcy
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,
                                test_size=0.30,random_state=i)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    ypred=lr.predict(x_test)
    acc=accuracy_score(y_test,ypred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print("Best accuracy of",maxacc,"at random state",maxrs)
```

BUILDING TRAIN TEST SPLIT

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                test_size=0.30,random_state=61)
```

```python
lr=LogisticRegression()
rfc=RandomForestClassifier()

etc=ExtraTreesClassifier()
bc=BaggingClassifier()
knn=KNeighborsClassifier()
svc=SVC()
gbc=GradientBoostingClassifier()

model=[lr,rfc,etc,bc,knn,svc,gbc]

```

```python
    #CHECKING ACCURACY SCORE OF MODELS
for m in model:  
     print(m)
     print('Accuracy score is',accuracy_score(y_test,ypred))
     print('confusion matrix is',confusion_matrix(y_test,ypred))
     print('report is',classification_report(y_test,ypred))
     print("-"*60)
```

CHECKING THE VALIDATION SCORE

```python
for j in model:
    print(j)
    score=cross_val_score(j,x,y,cv=5)
    print(score)
    print(score.mean())
    print("-"*60)
```

From above we notice that ExtraTrees Classifier giving best accuracy score,hence choosing

HYPERTUNING EXTARTREES

```python
from sklearn.model_selection import GridSearchCV
params={'n_estimators':[ 50, 120],
       'max_depth':[10,15,20],
       'n_estimators':[50,100,200,300],
    'criterion':['gini', 'entropy']}
gcv=GridSearchCV(etc,params,cv=5)
gcv.fit(x_train,y_train)
gcv.best_params_
```

```python
#Building model again using above parameters
etcmodel=ExtraTreesClassifier(criterion= 'entropy', max_depth= 20, n_estimators= 300)
etcmodel.fit(x_train,y_train)
ypred=etcmodel.predict(x_test)
print('Accuracy score is',accuracy_score(y_test,ypred))
print('confusion matrix is',confusion_matrix(y_test,ypred))
print('report is',classification_report(y_test,ypred))
```

```python
from sklearn import metrics as m
fpr,tpr,threshold=m.roc_curve(y_test,ypred)
roc_auc=m.auc(fpr,tpr)
display=m.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=etcmodel)
```
