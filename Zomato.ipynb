```python
exchange_rates = {
    'Indian Rupees(Rs.)': 0.014,  
    'Dollar($)': 1.0, 
    'Brazilian Real(R$)': 0.26, 
    'Indonesian Rupiah(IDR)': 7.0e-5,  
    'NewZealand($)': 0.72, 
    'Botswana Pula(P)': 0.088,  
    'Qatari Rial(QR)': 0.27,  
    'Rand(R)': 0.066,  
    'Sri Lankan Rupee(LKR)': 0.005,  
    'Turkish Lira(TL)': 0.12,  
    'Emirati Diram(AED)': 0.27,  
    'Pounds(\x8c£)': 1.31  
}

```

```python
# Function to convert currency to USD
def convert_to_usd(row):
    currency = row['Currency']  # Assuming 'Currency' is the column name
    amount = row['Average Cost for two']  # Assuming 'Amount' is the column name
    if currency == 'Dollar($)':
        return amount
    elif currency in exchange_rates:
        return amount * exchange_rates[currency]
    else:
        return None  # Return None for currencies without exchange rates

# Apply the conversion function to create a new column 'Amount (USD)'
df['Average Amount (USD)'] = df.apply(convert_to_usd, axis=1)
```

### MACHINE LEARNING

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor,BaggingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import SGDRegressor

```

```python
#Making object of ecah model
lrr=LinearRegression()
rfr=RandomForestRegressor()
etr=ExtraTreesRegressor()
gbr=GradientBoostingRegressor()
br=BaggingRegressor()
svr=SVR()
knnr=KNeighborsRegressor()
sgdr=SGDRegressor()
  


models1=[lrr,rfr,etr,gbr,br,svr,knnr,sgdr]
```

###### CHOOSING RANDOM STATE

```python
maxacc=0
maxrs=0
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    lrr.fit(x_train,y_train)
    ypred=lrr.predict(x_test)
    acc=r2_score(y_test,ypred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('Max accuacy',maxacc,'at random state',i)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=199)
```

## MACHINE LEARNING

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier,BaggingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
import xgboost
```

```python
#Making object of ecah model
lr=LogisticRegression()
rfc=RandomForestClassifier()
etc=ExtraTreesClassifier()
gbc=GradientBoostingClassifier()
bc=BaggingClassifier()
svc=SVC()
knn=KNeighborsClassifier()
sgd=SGDClassifier()
   


models=[lr,rfc,etc,gbc,bc,svc,knn,sgd]
```

###### CHOOSING RANDOM STATE

```python
maxacc=0
maxrs=0
for i in range(200):
    x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.30,random_state=i)
    lr.fit(x1_train,y1_train)
    ypred1=lr.predict(x1_test)
    acc=accuracy_score(y1_test,ypred1)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('Max accuacy',maxacc,'at random state',i)
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.30,random_state=199)
```

###### ACCUARCY SCORE FOR ALL MODELS

```python
for m in models:
    print("For",m)
    m.fit(x1_train,y1_train)
    ypred1=m.predict(x1_test)
    acc=accuracy_score(y1_test,ypred1)
    print(acc)
    print(confusion_matrix(y1_test,ypred1))
    print(classification_report(y1_test,ypred1))
    
    print("-"*60)
```

From above the best accuracy score is of ExtraTreesclassifier,RandomForestClassifier, model

###### CROSS VALIDATION

```python
for c in models:
    print("For",c)
    score1=cross_val_score(c,x1,y1,cv=5)
    print(score1.mean())
    print("-"*60)
```

From above cross validation sscore ,the Extartreesclassifier providing best accuracy score
