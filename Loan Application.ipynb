MACHINE LEARNING

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier,AdaBoostClassifier
import xgboost as xgb

```

```python
rfc=RandomForestClassifier()
etc=ExtraTreesClassifier()
lr=LogisticRegression()
knn=KNeighborsClassifier()
bc=BaggingClassifier()
abc=AdaBoostClassifier()
xg=xgb.XGBClassifier()


model=[rfc,etc,lr,knn,bc,abc,xg]
```

CHOOSING RANDOM STATE

```python
maxacc=0
maxrs=0
for i in range(200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train,y_train)
    ypred=lr.predict(x_test)
    acc=accuracy_score(y_test,ypred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('Max accuarcy ',maxacc,'at random state',maxrs)
```

CHOOSING RANDOM STATE=99 FOR MAX ACCURACY

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=99,test_size=0.20)
```

    ACCURACY SCORE FOR ALL MODELS

```python
for m in model:
    print("For",m,'model')
    m.fit(x_train,y_train)
    ypred=m.predict(x_test)
    acc=accuracy_score(y_test,ypred)
    print('Accuracy score',acc)
    print('Confusin Matrix',confusion_matrix(y_test,ypred))
    print('classification Report',classification_report(y_test,ypred))
    print("--"*60)
```

From above highest accuracy is of Random Forest Classifier

CROSS VALIDATION

```python
for c in model:
    print("For",c,'model')
    score=cross_val_score(c,x,y,cv=5)
    
    print(score.mean())
    print('Difference',(accuracy_score(y_test,ypred)-score.mean()))
   
    print("--"*60)
```

FROM ABOVE W ECONCLUDE THAT  
RANDOM FOREST CLASSIFIER PROVIDING HIGHEST ACCURACY WITH LEAST ERROR  

FINAL MODEL IS RANDOM FOREST CLASSIFIER

HYPER PARAMETER TUNING THE MODEL

```python
from sklearn.model_selection import GridSearchCV
parameters={ 'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt'],
    'bootstrap': [True, False] }

gcv=GridSearchCV(rfc,parameters)
gcv.fit(x_train,y_train)
gcv.best_params_
```

AGAIN BUILDING THE MODEL WITH THESSE PARAMETERS

```python
rfcmodel=RandomForestClassifier(bootstrap=False,
 max_depth= 10,
 max_features='sqrt',
 min_samples_leaf= 1,
 min_samples_split=10,
 n_estimators= 300)
rfcmodel.fit(x_train,y_train)
ypred=rfcmodel.predict(x_test)
acc=accuracy_score(y_test,ypred)
print('Accuracy',acc)
```

    AUC-ROC CURVE

```python
from sklearn import metrics as m
fpr,tpr,thresholds=m.roc_curve(y_test,ypred)
roc_auc=m.auc(fpr,tpr)
display=m.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=rfcmodel)
display.plot()
```

From above Auc-Roc curve,we observe that AUC score=0.81,which is Area under Curve which is good score as it is close to 1
